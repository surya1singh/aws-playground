```python
import boto3
import os
import json
import base64
```

```python
bedrock = boto3.client("bedrock-runtime", region_name=os.getenv("AWS_REGION", "us-east-1"))
```

```python
prompt = "generate a image where a rabbit is playing cricket"
```

```python
model_input = {
            "taskType": "TEXT_IMAGE",
            "textToImageParams": {"text": prompt},
            "imageGenerationConfig": {
                "width": 1024,
                "height": 1024,
                "numberOfImages": 1
            }
        }
```

```python
br = bedrock.invoke_model(
            modelId='amazon.titan-image-generator-v2:0',
            body=json.dumps(model_input),
            accept="application/json",
            contentType="application/json",
        )

raw = br["body"].read()
out = json.loads(raw)


b64 = None
if isinstance(out, dict):
    if "images" in out and out["images"]:
        b64 = out["images"][0]
    elif "artifacts" in out and out["artifacts"]:
        b64 = out["artifacts"][0].get("base64")
    elif "image" in out:
        b64 = out["image"]

if not b64:
    print(500, {"error": "Could not parse image from model response", "rawKeys": list(out.keys())})

img_bytes = base64.b64decode(b64)

key = f"image.png"
```

```python
with open(key,'wb') as fp:
    fp.write(img_bytes)
```
